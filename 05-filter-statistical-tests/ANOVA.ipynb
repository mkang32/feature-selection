### Definition & Assumptions

* Definition 

  * ANOVA tests the hypothesis that 2 or more samples have the same mean 

* Assumptions about the samples:

  * Samples are independent 
  * Samples are normally distributed 
  * Homogeneity of variance 

* If assumptions are not met, the results of ANOVA should be taken with caution. 

  

### Steps 

#### when categorical target variable and continuous independent variables

Classification with 2 target categories 

* Split data into 2 groups;  y=0, y=1.

|Var1 (y=0)|Var1 (y=1)|
| ----     |      ----|
|obs1      |obs 5     |
|obs2      |obs 6     |
|obs3      |obs 7     |
|...       |...       |
|Mean0     |Mean1     |
|Variance0 |Variance1 |

* Calculate the **great mean** = mean of all the observations

* Calculate the **model sum of squares** 
  * $$ \sum \text{# of obs} \times (\text{mean}- \text{great mean})^2 $$
  * $$ = \text{# of obs(0)} \times (\text{mean0} - \text{great mean})^2 + \text{# of obs(1)} \times (\text{mean1} - \text{great mean})^2$$ 

* Calculate the **residual sum of squares** 
  * $$ \sum(\text{obs}(0)-\text{mean0})^2 + \sum(\text{obs}(1)-\text{mean1})^2$$
* Calculate the **mean squares model** 
  * $$ \frac{\text{model sum of squares}}{\text{number of categories} - 1} $$
* Calculate the **mean squares error** 
  * $$ \frac{\text{Residual sum of squares}}{\text{total number of observation} - 1} $$
* Calculate F statistic 
  * $$ \text{F-statistic} = \frac{\text{mean squares model}}{\text{mean squares error}} $$
